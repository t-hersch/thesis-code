```python
from itertools import permutations

def sym_prod(l):
  perms = list(set(permutations(l)))
  total = Elt([{}])
  for i, perm in enumerate(perms):
    total += concatenationProductMany(list(perm))
  #return total
  return (1/len(perms))*total
```
